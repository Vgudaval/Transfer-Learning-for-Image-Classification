```python
import keras
import tensorflow as tf
import cv2
import os
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from keras.applications.vgg16 import VGG16
from keras.applications import EfficientNetB0
from keras.callbacks import EarlyStopping
from sklearn.metrics import f1_score, precision_score, recall_score, classification_report

import warnings
warnings.filterwarnings("ignore")
tf.get_logger().setLevel('ERROR')
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train = tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/DSCI_552_Project/Data/images',
    labels='inferred',
    label_mode='categorical',
    class_names=None,
    color_mode='rgb',
    batch_size=None,
    image_size=(224, 224),
    seed=2022,
    validation_split=0.30,
    subset="training",
    interpolation='bilinear',
    follow_links=False,
    crop_to_aspect_ratio=False,
)


val = tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/DSCI_552_Project/Data/images',
    labels='inferred',
    label_mode='categorical',
    class_names=None,
    color_mode='rgb',
    batch_size=None,
    image_size=(224, 224),
    seed=2022,
    validation_split=0.30,
    subset="validation",
    interpolation='bilinear',
    follow_links=False,
    crop_to_aspect_ratio=False,
)
```

```python
test_size = int(len(val)*0.5)
test = val.take(test_size)
val = val.skip(test_size)
print('Train size:', len(train))
print('Validation size:', len(val))
print('Test size:', len(test))
train = train.batch(32)
val = val.batch(32)
test = test.batch(176)
```

```python
vgg_model = VGG16(weights='imagenet', include_top=False, input_shape = (224, 224, 3))
vgg_model.trainable = False
print(vgg_model.summary())
```

```python
efficientNet_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape = (224, 224, 3))
efficientNet_model.trainable = False
print(efficientNet_model.summary())
```

```python
image_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.1),
    tf.keras.layers.RandomContrast(0.1)
])
```

```python
model_1_vgg = tf.keras.Sequential([
    image_augmentation,
    vgg_model,
    tf.keras.layers.Flatten(),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(20, activation='softmax')
])

model_1_vgg.compile(optimizer='adam', loss=tf.keras.losses.CategoricalCrossentropy(), metrics=['accuracy'])
earlyStopping = EarlyStopping(monitor='val_accuracy', mode='max', patience=10, restore_best_weights=True)

model_history_vgg = model_1_vgg.fit(train, epochs=100, validation_data = val, callbacks=[earlyStopping])
```

```python
print('Classification Report - VGG16')

for test_ in test:
    test_features, test_labels = test_
    test_labels_predicted = tf.one_hot(tf.argmax(model_1_vgg.predict(test_features), axis = 1), depth = 20)
    test_labels_predicted = test_labels_predicted.numpy()
    test_labels = test_labels.numpy()
    print(classification_report(test_labels, test_labels_predicted))
```

```python
plt.plot(model_history_vgg.history['loss'])
plt.plot(model_history_vgg.history['val_loss'])
plt.title('Training and Validation errors vs. epochs - VGG16')
plt.ylabel('Error')
plt.xlabel('Epoch')
plt.legend(['train', 'val'], loc='auto')
plt.show()
```

```python
model_eff_net = tf.keras.Sequential([
    image_augmentation,
    efficientNet_model,
    tf.keras.layers.Flatten(),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(20, activation='softmax')
])

model_eff_net.compile(optimizer='adam', loss=tf.keras.losses.CategoricalCrossentropy(), metrics=['accuracy'])
early_stopping_en = EarlyStopping(monitor='val_accuracy', mode='max', patience=10, restore_best_weights=True)

model_history_EN = model_eff_net.fit(train, epochs=100, validation_data = val, callbacks=[early_stopping_en])
```

```python
print('Classification Report - EfficientNetB0')

for t in test:
    t_feat, t_label = t
    test_labels_pred_en = tf.one_hot(tf.argmax(model_eff_net.predict(t_feat), axis = 1), depth = 20)
    test_labels_pred_en = test_labels_pred_en.numpy()
    test_labels_true = t_label.numpy()
    print(classification_report(test_labels_true, test_labels_pred_en))
```

```python
plt.plot(model_history_EN.history['loss'])
plt.plot(model_history_EN.history['val_loss'])
plt.title('Training and Validation errors vs. epochs - EfficientNetB0')
plt.ylabel('Error')
plt.xlabel('Epoch')
plt.legend(['train', 'val'], loc='auto')
plt.show()
```
